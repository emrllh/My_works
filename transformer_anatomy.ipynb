The code you provided calculates the attention weights using the softmax function and then sums these weights along the last axis (axis=-1).

Reasoning for using axis=-1

Understanding the Shape: The weights tensor likely has a shape like (batch_size, sequence_length, sequence_length). Each element weights[i, j, k] represents the attention weight between the i-th element in the batch, the j-th word in the query sequence, and the k-th word in the key sequence.

Summing across the Key Dimension: We want to sum the attention weights across all possible key words for each query word. This is done to normalize the attention weights so that they represent a probability distribution over the key words. Since the key words are represented by the last dimension of the weights tensor, we use axis=-1 to sum across that dimension.

In simpler terms:

Imagine you have a table where rows represent query words, columns represent key words, and the values in the cells are attention weights. Using axis=-1 is like summing up the values in each row of the table. This gives you the total attention weight for each query word, considering all the key words.

Why it's Important:

Normalization: The sum of attention weights along axis=-1 should ideally be close to 1 for each query word. This ensures that the weights represent a probability distribution and that the attention mechanism is properly normalized.
Focus: This summation helps the model focus on the most relevant key words for each query word by assigning higher attention weights to them. The weights after summation represent the relative importance of each key word to the query word

```python
#Put everyting in a function

def scaled_dot_product_attention(query, key, value):
  dim_k = query.shape[-1]
  scores = tf.matmul(query, key, transpose_b=True)/sqrt(dim_k)
  weights = tf.nn.softmax(scores, axis=-1)
  return tf.matmul(weights, value)
```

#### *Multi-headed attention*

```python
class AttentionHead(tf.keras.layers.Layer):
  def __init__(self,embed_dim, head_dim):
    super().__init__()

    self.q = tf.keras.layers.Dense(head_dim,activation='relu')
    self.k = tf.keras.layers.Dense(head_dim,activation='relu')
    self.v = tf.keras.layers.Dense(head_dim,activation='relu')

  def call(self, hidden_state):
    attn_outputs = scaled_dot_product_attention(
        self.q(hidden_state), self.k(hidden_state), self.v(hidden_state))
    return attn_outputs
```

```python
class MultiHeadAttention(keras.layers.Layer):
  def __init__(self, config):
    super().__init__()
    embed_dim = config.hidden_size
    num_heads = config.num_attention_heads
    head_dim = embed_dim // num_heads
    self.heads =keras.layers.Dense(
        [AttentionHead(embed_dim,head_dim) for _ in range(num_heads)])

    self.output_linear = keras.layers.Dense(embed_dim,activation='relu')


  def call(self, hidden_state):
    x = keras.layers.concatenate([h(hidden_state) for h in self.heads])
    x = self.output_linear(x)
    return x
```

### The Feed-Forward Layer

```python
import keras
class FeedForward(keras.layers.Layer):
  def __init__(self, config):
    super().__init__()
    self.linear_1 = keras.layers.Dense(config.hidden_size, config.intermediate_size)
    self.linear_2 = keras.layers.Dense(config.intermediate_size, config.hidden_size)
    self.gelu = keras.layers.Activation("gelu")
    self.dropout = keras.layers.Dropout(config.hidden_dropout_prob)

    def call(self, x):
      x = self.linear_1(x)
      x = self.gelu(x)
      x = self.linear_2(x)
      x = self.dropout(x)
      return x
```

### Adding Layer Normalization

```python
class TransformerEncoderLayer(keras.layers.Layer):
  def __init__(self, config):
    super().__init__()

    self.layer_norm_1 = keras.layers.LayerNormalization(config.hidden_size)
    self.layer_norm_2 = keras.layers.LayerNormalization(config.hidden_size)
    self.attention = MultiHeadAttention(config)
    self.feed_forward = FeedForward(config)

  def call(self, x):
    # Apply layer normalization and then copy input into query, key, value
    hidden_state = self.layer_norm_1(x)
    # Apply attention with a skip connection
    x = x + self.attention(hidden_state)
    # Apply feed-forward layer with a skip connection
    x = x + self.feed_forward(self.layer_norm_2(x))

    return x

```

What is a Skip Connection?

In deep learning, a skip connection (or residual connection) is a technique where the output of a layer is added to the input of a later layer. Instead of just passing the output of one layer directly to the next, we create a "shortcut" where the original input is also combined with the output.

How it Works in Your Code:

x: Represents the input to the current layer (likely the output of a previous layer).
self.feed_forward(self.layer_norm_2(x)): This part applies a feed-forward network and layer normalization to the input x. Let's call the result of this operation F(x).
x + F(x): This is where the skip connection happens. The original input x is added to the transformed output F(x). The result is then assigned back to x, becoming the output of this layer.
Why Skip Connections Are Used:

Improved Gradient Flow: Skip connections help gradients flow more easily through deep networks during training. This is especially important in very deep architectures, where gradients can vanish or explode, making it difficult to train effectively.
Faster Convergence: By allowing gradients to flow more freely, skip connections can help the network converge to a good solution faster.
Mitigation of Vanishing Gradients: In deep networks, gradients can become very small as they are backpropagated through many layers. Skip connections provide an alternate path for gradients to flow, mitigating this problem.
Enables Deeper Networks: Skip connections have been a key factor in enabling the training of very deep neural networks.
In the Context of Transformers:

Skip connections are a fundamental component of Transformer architectures. They are used in both the encoder and decoder parts to help stabilize training and improve performance. They are crucial for the success of Transformers in various natural language processing tasks.

### Positional Embeddings

```python
class Embeddings(keras.layers.Layer):
  def __init__(self, config):
    super().__init__()
    self.token_embeddins = keras.layers.layer.Embedding(config.vocab_size,
                                                        config.hidden_size)
    self.position_embeddings = keras.layers.layer.Embedding(config.max_position_embeddings,
                                                            config.hidden_size)
    self.layer_norm = keras.layers.LayerNormalization(config.hidden_size, eps=1e-12)
    self.dropout = keras.layers.Dropout()

  def call(self, input_ids):
    # Create position IDs for input sequence
    seq_length = input_ids.shape[1]
    position_ids = tf.range(seq_length)

    # Create token and position embeddings
    token_embeddings = self.token_embeddings(input_ids)
    position_embeddings = self.position_embeddings(position_ids)

    # Combine token and position embeddings
    embeddings = token_embeddings + position_embeddings
    embeddings = self.layer_norm(embeddings)
    embeddings = self.dropout(embeddings)
    return embeddings
```

```python
embedding_layer = Embeddings(config)
embedding_layer(inputs.input_ids).shape
```

```python
# Put them all together to create Transformer Encoder combining embeddings with the encoder layers
class TransformerEncoder(keras.layers.layer):
  def __init__(self, config):
    super.__init__()
    self.embeddings = Embeddings(config)
    self.layers = [TransformerEncoderLayer(config) for _ in range(config.num_hidden_layers)]

  def call(self, x):
    x = self.embeddings(x)
    for layer in self.layers:
      x = layer(x)
    return x
```

#### Adding a Classification Head

```python
class TransformerForSequenceClassification(keras.layers.layer):
  def __init__(self,config):
    super().__init__()
    self.encoder = TransformerEncoder(config)
    self.dropout = keras.layers.layer(config.hidden_size)
    self.classifier = keras.layers.layer(config.num_labels)

  def call(self, x):
    x = self.encoder(x)[:, 0, :] # select hidden state of [CLS] token
    x = self.dropout(x)
    x = self.classifier(x)
    return x
```
